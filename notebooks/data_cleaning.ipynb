```python
from pathlib import Path
import pandas as pd

data_file = Path('..') / 'data' / 'processed' / 'movies_1990_onwards.csv'

data = pd.read_csv(data_file)
data.info()
data.head()


```

```python
# Sort dataset by startYear
df = data.sort_values(by="startYear")

# Extract and display unique years (sorted)
unique_years = sorted(df["startYear"].dropna().unique())

# Display them
print(unique_years)
```

```python
df = df[df['startYear'] <= 2020]
unique_years = sorted(df["startYear"].dropna().unique())

# Display them
print(unique_years)
```

```python
df.head()
```

```python
# Convert startYear to integer (handling NaN safely)
df["startYear"] = df["startYear"].fillna(0).astype(int)

# Rename the column
df.rename(columns={"startYear": "releaseYear"}, inplace=True)

df.head()
```

```python
# Drop endYear column
df.drop(columns=["endYear"], inplace=True)
```

```python
# Drop unnecessary columns
df.drop(columns=["titleType", "genre_list"], inplace=True)

# Verify
print(df.columns)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["primaryTitle", "originalTitle"])
```

```python
df.isnull().sum()
```

```python
#fill missing values in 'genres' with 'Unknown'
df['genres'] = df['genres'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['genres_sorted'] = df['genres'].apply(lambda x: tuple(sorted(x)) if x != ['Unknown'] else 'Unknown')

median_per_genres = df.groupby('genres_sorted')['runtimeMinutes'].median()

overall_median = df['runtimeMinutes'].median()

def fill_runtime(row):
    if pd.notna(row['runtimeMinutes']):
        return row['runtimeMinutes']
    return median_per_genres.get(row['genres_sorted'], overall_median)

df['runtimeMinutes'] = df.apply(fill_runtime, axis=1)

df.drop(columns='genres_sorted', inplace=True)
```

```python
df.isnull().sum()
```

```python
missing_runtime = df[df['runtimeMinutes'].isna()]
missing_runtime.head()
```

```python
median_per_genres.isna().sum()
```

### There are some genres for which the grouped genre runtime median is 0 leading to 50 records still having runtimeMinutes as NaN. So, we will impute these records with overall median of runtimeMinutes

```python
df['runtimeMinutes'] = df['runtimeMinutes'].fillna(overall_median)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import numpy as np

df['genres_sorted'] = df['genres'].apply(lambda x: tuple(sorted(x)) if x != ['Unknown'] else 'Unknown')

group_medians = df.groupby(['releaseYear', 'genres_sorted'])[['averageRating', 'numVotes']].median().reset_index()

df = df.merge(group_medians, on=['releaseYear', 'genres_sorted'], how='left', suffixes=('', '_median'))

overall_medians = df[['averageRating', 'numVotes']].median()

df['averageRating'] = df['averageRating'].fillna(df['averageRating_median']).fillna(overall_medians['averageRating'])
df['numVotes'] = df['numVotes'].fillna(df['numVotes_median']).fillna(overall_medians['numVotes'])

df.drop(columns=['genres_sorted', 'averageRating_median', 'numVotes_median'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
data_encoded = pd.concat([df, df['genres'].str.get_dummies(sep=',')], axis=1)


data_encoded.drop(columns=['genres'], inplace=True)
```

```python
data_encoded.to_csv("data_final.csv", index=False, encoding='utf-8')
```

```python
import pandas as pd
import plotly.express as px

# List of genre columns (excluding Unknown if you want)
genre_cols = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography', 'Comedy',
              'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'Game-Show',
              'History', 'Horror', 'Music', 'Musical', 'Mystery', 'News',
              'Reality-TV', 'Romance', 'Sci-Fi', 'Short', 'Sport', 'Talk-Show',
              'Thriller', 'War', 'Western']

# Compute average runtime for each genre
avg_runtime_per_genre = {}

for genre in genre_cols:
    # Select only movies with this genre
    avg_runtime_per_genre[genre] = data_encoded.loc[data_encoded[genre] == 1, 'runtimeMinutes'].mean()

# Convert to DataFrame for plotting
avg_runtime_df = pd.DataFrame(list(avg_runtime_per_genre.items()), columns=['Genre', 'AvgRuntime'])
```
